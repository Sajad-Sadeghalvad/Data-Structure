```python
def factorial(num):
    if num == 0:
        return 1
    else: 
        return num * factorial(num - 1)
```

```python
factorial(5)
```

```python
def fibonacci(num):
    if num <= 1:
        return num
    return fibonacci(num - 2) + fibonacci(num - 1)
```

```python
fibonacci(7)
```

```python
#      محاسبه ضرب دو عدد
def multipliction(a, b):
    if b == 1:
        return a
    return multipliction(a, b - 1) + a
```

```python
multipliction(2, 3)
```

```python
#      محاسبه تقسیم صحیح دو عدد
def division(a, b):
    if a < b:
        return 0
    return division(a - b, b) + 1
```

```python
division(5, 10)
```

```python
division(9, 2)
```
