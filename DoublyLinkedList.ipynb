```python
class Node:
    def __init__(self, data):
        self.data = data
        self.Next = None
        self.Prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_first(self, data):
        new_node = Node(data)
        if self.head:
            self.head.Prev = new_node
            new_node.Next = self.head
        self.head = new_node

    def insert_last(self, data):
        if self.head is None:
            self.head = Node(data)
            return
        c = self.head
        while c.Next:
            c = c.Next
        new_node = Node(data)
        c.Next = new_node
        new_node.Prev = c

    def insert_after(self, value, data):
        c = self.head
        while c:
            if c.data == value:
                new_node = Node(data)
                new_node.Next = c.Next
                new_node.Prev = c
                if c.Next:
                    c.Next.Prev = new_node
                c.Next = new_node
                return
            c = c.Next
        print("Value not found")

    def insert_before(self, value, data):
        if self.head is None:
            print("List is empty")
            return
        if self.head.data == value:
            self.insert_first(data)
            return
        c = self.head.Next
        while c:
            if c.data == value:
                new_node = Node(data)
                new_node.Next = c
                new_node.Prev = c.Prev
                c.Prev.Next = new_node
                c.Prev = new_node
                return
            c = c.Next
        print("Value not found")

    def del_first(self):
        if self.head is None:
            print("List is empty")
            return
        temp = self.head
        self.head = temp.Next
        if self.head:
            self.head.Prev = None
        del temp

    def del_last(self):
        if self.head is None:
            print("List is empty")
            return
        if self.head.Next is None:
            self.del_first()
            return
        c = self.head
        while c.Next:
            c = c.Next
        c.Prev.Next = None
        del c

    def del_after(self, x):
        c = self.head
        while c:
            if c.data == x and c.Next:
                temp = c.Next
                c.Next = temp.Next
                if temp.Next:
                    temp.Next.Prev = c
                del temp
                return
            c = c.Next
        print("Invalid delete")

    def del_before(self, x):
        if self.head is None or self.head.data == x:
            print("Invalid delete")
            return
        c = self.head.Next
        while c:
            if c.data == x and c.Prev:
                temp = c.Prev
                if temp.Prev:
                    temp.Prev.Next = c
                    c.Prev = temp.Prev
                else:
                    self.head = c
                    c.Prev = None
                del temp
                return
            c = c.Next
        print("Invalid delete")

    def del_x(self, x):
        c = self.head
        while c:
            if c.data == x:
                if c.Prev:
                    c.Prev.Next = c.Next
                else:
                    self.head = c.Next
                if c.Next:
                    c.Next.Prev = c.Prev
                del c
                return
            c = c.Next
        print("Value not found")

    def del_all(self):
        while self.head:
            self.del_first()

    def display(self):
        c = self.head
        while c:
            print(c.data, end=" <-> ")
            c = c.Next
        print("None")
```

```python
def main():
    dll = DoublyLinkedList()

    dll.insert_first(10)
    dll.insert_last(20)
    dll.insert_last(30)
    dll.insert_after(20, 25)
    dll.insert_before(10, 5)

    dll.display()

    dll.del_first()
    dll.del_last()
    dll.del_x(20)
    dll.del_after(10)
    dll.del_before(10)

    dll.display()

    dll.del_all()
    dll.display()

if __name__ == "__main__":
    main()
```
