```python
#    با استفاده از فضای کمکی
def sort1(A):
    A = A[:]
    B = []
    while A:
        m = A[0]
        k = 0
        for i in range(1, len(A)):
            if A[i] < m:
                m = A[i]
                k = i
        B.append(m)
        A.pop(k)
    return B
```

```python
l = [10, 20 , -1, 100, 50, 33, -22, -34]
```

```python
sort1(l)
```

```python
def Bubble(A):
    for i in range(len(A) - 1):
        for j in range(len(A) - 1 - i):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
    return A
```

```python
Bubble(l)
```

```python
def Selection(A):
    for i in range(len(A) - 1):
        k = i
        for j in range(i + 1, len(A)):
            if A[j] < A[k]:
                k = j
        A[i], A[k] = A[k], A[i]
    return A
```

```python
Selection(l)
```

```python
def merge(A):
    if len(A) > 1:
        mid = len(A) // 2
        left = A[:mid]
        right = A[mid:]
        merge(left)
        merge(right)
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                A[k] = left[i]
                i += 1
            else:
                A[k] = right[j]
                j += 1
            k += 1
        while i < len(left):
            A[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            A[k] = right[j]
            j += 1
            k += 1
    return A
```

```python
merge(l)
```

```python
def Quick(A, low, high):
    if low < high:
        pi = partition(A, low, high)
        Quick(A, low, pi - 1)
        Quick(A, pi + 1, high)

def partition(A, low, high):
    pivot = A[high]
    i = low - 1
    for j in range(low, high):
        if A[j] < pivot:
            i += 1
            A[i], A[j] = A[j], A[i]

    A[i + 1], A[high] = A[high], A[i + 1]
    return i + 1
```

```python
A = [10, 7, 8, 9, 1, 5]
Quick(A, 0, len(A) - 1)
print(A)
```
