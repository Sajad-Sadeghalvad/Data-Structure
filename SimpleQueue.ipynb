```python
class Queue:
    def __init__(self, max=100):
        self.queue = [None] * max
        self.front = -1
        self.rear = -1
        self.max = max

    def enqueue(self, data):
        if self.rear >= self.max - 1:
            print('Queue is full')
            return
        if self.front == -1:
            self.front = 0
            self.rear = 0
            self.queue[0] = data
            return
        self.rear += 1
        self.queue[self.rear] = data

    def dequeue(self):
        if self.front == -1:
            print('Queue is empty')
            return None
        k = self.queue[self.front]
        if self.front == self.rear:
            self.front = -1
            self.rear = -1
        else:
            self.front += 1
        return k
```

```python
q = Queue(5)
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
```
