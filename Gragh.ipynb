```python
class Gragh:   # with matrix
    def __init__(self, nodes):
        self.size = nodes
        self.matrix = []
        for _ in range(nodes):
            self.matrix.append([0 for i in range(nodes)])

    def add_edge(self, vertex1, vertex2):
        self.matrix[vertex1][vertex2] = 1
        self.matrix[vertex2][vertex1] = 1

    def delete_edge(self, vertex1, vertex2):
        self.matrix[vertex1][vertex2] = 0
        self.matrix[vertex2][vertex1] = 0

    def print_matrix(self):
        for row in self.matrix:
            for j in row:
                print(j , end=" ")
            print()
```

```python
g = Gragh(4)
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.print_matrix()
```

```python
class Node:
    def __init__(self, data):
        self.vertex = data
        self.next = None

class GraphWithList:
    def __init__(self, node):
        self.V = node
        self.graph = [None] * self.V

    def add_edge(self, s, d):
        n = Node(d)
        n.next = self.graph[s]
        self.graph[s] = n
        m = Node(s)
        m.next = self.graph[d]
        self.graph[d] = m

    def show(self):
        for i in range(self.V):
            print(i, ": head", end=" ")
            t = self.graph[i]
            while t:
                print(f"-> {t.vertex}", end=" ")
                t = t.next
            print()
```

```python
q = GraphWithList(5)
q.add_edge(0, 1)
q.add_edge(0, 4)
q.add_edge(1, 2)
q.add_edge(1, 3)
q.add_edge(1, 4)
q.add_edge(2, 3)
q.add_edge(3, 4)

q.show()
```

```python
def BFS(gragh, start, List):
    q = []
    List.append(start)
    q.append(start)
    while q:
        m = q.pop(0)
        print(m)
        for n in gragh[m]:
            if n not in List:
                List.append(n)
                q.append(n)

g = {
    '0': ['1', '2'],
    '1': ['3', '4', '6'],
    '2': ['5', '6'],
    '3': [ ],
    '4': [ ],
    '5': [ ],
    '6': [ ]
}

BFS(g, '0', List= [])
```

```python
def DFS(graph, start, List):
    if start not in List:
        print(start)
        List.append(start)
        for n in graph[start]:
            DFS(graph, n, List)
```

```python
DFS(g, '0', [])
```
