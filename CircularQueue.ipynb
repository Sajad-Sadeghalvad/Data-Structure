```python
class CircularQueue:
    def __init__(self, size):
        self.rear = -1
        self.front = -1
        self.Queue = [None] * size
        self.limit = size

    def is_empty(self):
        return self.rear == -1
    
    def is_full(self):
        return (self.rear + 1) % self.limit == self.front
    
    def enqueue(self, data):
        if self.is_full():
            print('Queue is full')
            return -1
        if self.is_empty():
            self.front = 0
            self.rear = 0
        else:
            self.rear = (self.rear + 1) % self.limit
        self.Queue[self.rear] = data

    def dequeue(self):
        if self.is_empty():
            print('Queue is empty')
            return -1
        removed = self.Queue[self.front]
        if self.front == self.rear:
            self.front = -1
            self.rear = -1
        else:
            self.front = (self.front + 1) % self.limit
        return removed
    
    def display(self):
        if self.is_empty():
            print('Queue is empty')
            return
        print('Circular Queue:')
        i = self.front
        while True:
            print(self.Queue[i], end= " ")
            if i == self.rear:
                break
            i = (i + 1) % self.limit
        print()

    def show_valid(self):
        if self.is_empty():
            print("Queue is empty")
            return
        if self.rear >= self.front:
            for i in range(self.front, self.rear + 1, 1):
                print(self.Queue[i])
            else:
                for i in range(self.front , self.limit, 1):
                    print(self.Queue[i])
                for i in range(0, self.rear + 1, 1):
                    print(self.Queue[i])

    def show_invalid(self):
        for i in range(self.limit):
            if self.Queue[i] is None:
                print(i)

    def find(self, data):
        if self.is_empty():
            return -1
        i = self.front
        while True:
            if self.Queue[i] == data:
                return i
            if i == self.rear:
                break
            i = (i + 1) % self.limit

        return -1
    
    def replace(self, old, new):
        index = self.find(old)
        if index == -1:
            return False
        self.Queue[index] = new
        return True
```

```python
def main():
    c = CircularQueue(4)
    c.enqueue(30)
    c.enqueue(10)
    c.enqueue(50)
    c.display()
    c.enqueue(40)
    c.enqueue(12)
    c.display()
    print(c.find(50))
    print(c.find(70))
    # c.dequeue()
    c.display()
    c.replace(50, 71)
    c.display()
if __name__ == "__main__":
    main()
```
