```python
class Node:
    def __init__(self, data):
        self.Data = data
        self.Next = None
    
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_first(self, data):
        new_node = Node(data)
        new_node.Next = self.head
        self.head = new_node

    def insert_last(self, data):
        new_node = Node(data)
        if self.head is None:
            print("Creating a Singly Linked List...")
            self.head = new_node
            return 
        c = self.head
        while c.Next:
            c = c.Next
        c.Next = new_node

    def insert_after(self, value, new_data):
        if self.head is None:
            print('List is Empty...')
            return
        c = self.head
        while c:
            if c.Data == value:
                new_node = Node(new_data)
                new_node.Next = c.Next
                c.Next = new_node
                return
            c = c.Next
        print('Not Found')    

    def insert_before(self, value, data):
        if self.head is None:
            print('List is Empty...')
            return
        if self.head.Data == value:
            self.insert_first(data)
            return
        c = self.head
        while c.Next:
            if c.Next.Data == value:
                new_node = Node(data)
                new_node.Next = c.Next
                c.Next = new_node
                return
            c = c.Next
        print('not found')

    def display(self):
        if self.head is None:
            print('List is Empty')
            return
        c = self.head
        while c:
            print(c.Data, end=" -> ")
            c = c.Next
        print("None")

    def del_first(self):
        if self.head is None:
            print("List is Empty")
            return
        c = self.head
        self.head = c.Next
        del c

    def del_last(self):
        if self.head is None:
            print('List is Empty')
            return
        if self.head.Next is None:
            self.del_first()
            return
        c = self.head
        while c.Next.Next :
            c = c.Next
        del c.Next
        c.Next = None

    def del_after(self, value):
        if self.head is None:
            print('List is Empty')
            return 
        if self.head.Next is None:
            print('error1')
            return
        if self.head.Data == value:
            a = self.head.Next
            self.head.Next = a.Next
            del a
            return
        c = self.head
        while c.Next:
            if c.Data == value:
                a = c.Next
                c.Next = a.Next
                del a
                return
            c = c.Next
        print('Value not found')

    def del_before(self, value):
        if self.head is None:
            print("List is Empty")
            return
        if self.head.Next is None:
            print("only one value in List")
            return 
        if self.head.Data == value:
            print('error')
            return
        if self.head.Next.Data == value:
            self.del_first()
            return
        if self.head.Next.Next is None:
            print('error')
            return
        c = self.head
        while c.Next.Next:
            if c.Next.Next.Data == value:
                a = c.Next
                c.Next = a.Next
                del a
                return
            c = c.Next
        print('value not found')
```
