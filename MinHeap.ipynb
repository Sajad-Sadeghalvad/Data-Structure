```python
def min_heapify(A, i):
    smallest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < len(A) and A[l] < A[smallest]:
        smallest = l
    if r < len(A) and A[r] < A[smallest]:
        smallest = r

    if smallest != i:
        A[i], A[smallest] = A[smallest], A[i]
        min_heapify(A, smallest)


def build_min_heap(A):
    for i in range(len(A) // 2 - 1, -1, -1):
        min_heapify(A, i)


def insert(A, value):
    A.append(value)
    i = len(A) - 1

    while i > 0:
        parent = (i - 1) // 2
        if A[parent] > A[i]:
            A[parent], A[i] = A[i], A[parent]
            i = parent
        else:
            break


def delete(A):
    if len(A) == 0:
        return None

    root = A[0]
    A[0] = A[-1]
    A.pop()

    if len(A) > 0:
        min_heapify(A, 0)

    return root
```

```python
A = [9, 5, 6, 2, 3]
build_min_heap(A)
print(A)
```

```python
insert(A, 1)
print(A)
```
