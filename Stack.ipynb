```python
class Stack:
    def __init__(self, limit=10):
        self.stack = []
        self.limit = limit

    def push(self, data):
        if len(self.stack) >= self.limit:
            print("Stack is full")
            return -1
        self.stack.append(data)

    def pop(self):
        if len(self.stack) == 0:
            print("Stack is empty")
            return -1
        return self.stack.pop()

    def peek(self):
        if len(self.stack) == 0:
            print("Stack is empty")
            return -1
        return self.stack[-1]

    def find(self, value):
        for i in range(len(self.stack)):
            if self.stack[i] == value:
                print(i)

    def find_first(self, value):
        for i in range(len(self.stack)):
            if self.stack[i] == value:
                print(i)
                return

    def find_of_last(self, value):
        for i in range(len(self.stack) - 1, -1, -1):
            if self.stack[i] == value:
                print(i)
                return

    def find_in_list(self, value):
        result = []
        for i in range(len(self.stack)):
            if self.stack[i] == value:
                result.append(i)
        print(result)

    def replace(self, old, new):
        for i in range(len(self.stack)):
            if self.stack[i] == old:
                self.stack[i] = new

    def is_full(self):
        return len(self.stack) >= self.limit

    def is_empty(self):
        return len(self.stack) == 0

    def show(self):
        return self.stack

    def replace_nth_x(self, n, old, new):
        if n <= 0:
            return False
        count = 0
        for i in range(len(self.stack)):
            if self.stack[i] == old:
                count += 1
                if count == n:
                    self.stack[i] = new
                    return
        print("Not found")
```

```python
s = Stack(limit=5)

s.push(10)
s.push(20)
s.push(30)
s.push(20)
s.push(40)
s.push(50)
```

```python
s.peek()
```

```python
s.replace(20, 99)
```
